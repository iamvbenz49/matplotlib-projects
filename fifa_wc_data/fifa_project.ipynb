```python
#loading necessary packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fifa = pd.read_csv("fifa_data.csv")
fifa.head(5)
```

# **Histograms**

```python
bins = [40,50,60,70,80,90]

plt.hist(fifa.Overall,bins=bins,color="#2f5")

plt.title("SKills vs NUmber of Players")
plt.xlabel("Skill Level")
plt.ylabel("NUmber of Players")

plt.show()
```
