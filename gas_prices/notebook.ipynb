# **Load Necessary libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Real World Examples**

```python
gas = pd.read_csv("gas_prices.csv")

gas
```

```python
plot = plt.figure(figsize=(15,10))


plt.title("Gas Prices($USD)")

# plt.plot(gas.Year,gas.USA,'b.-')
# plt.plot(gas.Year,gas.Australia,'r.-')
# plt.plot(gas.Year,gas.Germany,'g.-')
# plt.plot(gas.Year,gas.Italy,'m.-')

for country in gas:
  if country!="Year":
    plt.plot(gas.Year,gas[country],'.-')

plt.legend([i for i in gas if i!="Year"])
plt.xticks(gas.Year)

plt.xlabel("Year")
plt.ylabel("Price")

plt.savefig("gasPrices.png",dpi=300)
plt.show()
```
